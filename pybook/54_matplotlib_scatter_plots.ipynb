## Creating the first scatter plot for variables `x` and `y`: `plt.scatter(x, y, s=, color=, edgecolor=, linewidth=, alpha=)`

Scatter plots are great when we want to plot two sets of values together, and see if they are correlated.

`plt.scatter()` API: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter

Let's just plot 2 variables: `x` and `y` to see if there is a correlation between them:

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)     # 's=' => marker size, 'c=' => marker color

plt.tight_layout()
```

## Adding colors to a scatter plot on a 'per marker basis', with a 'colormap' & 'colorbar': `plt.scatter(c=<array-like>, cmap=)` arguments, & `plt.colorbar().set_label(<value>)`

Having the ability to add multiple colors (and/or sizes) allows us to add additional dimensions to our plot.

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)     # 'color=colors'

plt.tight_layout()
```

Now the marker colors above are just different shades of grey. We might want to use a different **colormap**, and add a corresponding **colorbar**:

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)     # Added colormap with `cmap=`

# Add colorbar and set its label
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
```

## Adding marker sizes on a 'per marker basis': `plt.scatter(s=<array-like>)` argument

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)     # `s=sizes`

# Add colorbar and set its label
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
```

## Let's plot some real world data for a YouTube video view count and likes

```python
plt.style.use('seaborn')

# Let's read the csv file that contains 3 columns: view_count, likes, ratio (likes/dislikes). Each row represents a video.
data = pd.read_csv('work_directory/matplotlib/code_snippets/07-ScatterPlots/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, s=50, c='lightblue', edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

### Using a log-scale for the scatter plot: `plt.xscale('log')` & `plt.yscale('log')`

Let's use a **log-scale** to see the correlation better:

```python
plt.style.use('seaborn')

# Let's read the csv file that contains 3 columns: view_count, likes, ratio. Each row represents a video.
data = pd.read_csv('work_directory/matplotlib/code_snippets/07-ScatterPlots/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, s=50, c='lightblue', edgecolor='black', linewidth=1, alpha=0.75)

# Changing x and y axis scales to logarithmic 
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

Now let's **add the ratio (of likes/dislikes) as a color** to this scatter plot:

```python
plt.style.use('seaborn')

# Let's read the csv file that contains 3 columns: view_count, likes, ratio. Each row represents a video.
data = pd.read_csv('work_directory/matplotlib/code_snippets/07-ScatterPlots/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, s=50, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)   # 'c=ratio', 'cmap='

# Add colorbar and set its label
cbar= plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# Changing x and y axis scales to logarithmic 
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

Since the dataset above is extracted from YouTube's trending videos, you would see that **most of the likes/dislikes ratios are pretty high.**

However, we do see a **few relatively low ratios (50-60%)**. Now we could look into those specific videos to understand why their ratios stand out on the low side. Are these videos on subjects where the world view is very polarized? Are they of a political/religious nature? Is there any correlation between ratios and the content category?
