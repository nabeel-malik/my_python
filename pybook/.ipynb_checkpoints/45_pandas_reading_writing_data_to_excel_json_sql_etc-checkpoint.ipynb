## Read/write `.csv` files

### Read from `.csv`: `pandas.read_csv()` method

```python
import pandas as pd
```

```python
df = pd.read_csv('work_directory/pandas/data/survey_results_public.csv', index_col='Respondent')
df_schema = pd.read_csv('work_directory/pandas/data/survey_results_schema.csv', index_col='Column')
```

```python
pd.set_option('display.max_rows', 85)
pd.set_option('display.max_columns', 85)
```

```python
df.head()
```

### Write to `.csv`: `DataFrame.to_csv()` method

```python
filt_india = df['Country'] == 'India'
df_india = df.loc[filt_india]
```

```python
df_india
```

```python
df_india.to_csv('work_directory/pandas/output/survey_results_public_india.csv')
```

We can also **write tab-delimited files**. These are almost exactly the same as `.csv` files, but instead of the data being separated by commas (','), the data is instead separated by tabs ('\t').

To write tab-delimited files we will still use the `DataFrame.to_csv()` method, but with the `sep='\t'` argument.

```python
df_india.to_csv('work_directory/pandas/output/survey_results_public_india.tsv', sep='\t')
```

To **read tab-delimited files**, we will just add the `sep='\t'` argument to `pandas.read_csv()` method.

## Read/write Excel files

### Write to Excel file: `DataFrame.to_excel()` method

Firstly make sure that you have run `pip install xlwt openpyxl xlrd` in Anaconda prompt.

```python
df_india.to_excel('work_directory/pandas/output/survey_results_public_india.xlsx')
```

### Read from Excel file: `pandas.read_excel()` method

```python
df_india_excel = pd.read_excel('work_directory/pandas/output/survey_results_public_india.xlsx', index_col='Respondent')
```

```python
df_india_excel.head()
```

## Read/write JSON files

### Write to JSON file: `DataFrame.to_json()` method

If we write the JSON file using the `DataFrame.to_json()` method with the detault arguments, the JSON file written will be **dictionary-like**:

```python
df_india.to_json('work_directory/pandas/output/survey_results_public_india_dict-like.json')
```

If we want to write a JSON file that is **list-like** instead of the default *dictionary-like*, we can just add the `orient='records` and `lines=True` arguments.

```python
df_india.to_json('work_directory/pandas/output/survey_results_public_india_list-like.json', orient='records', lines=True)
```

### Read from JSON file: `pandas.read_json()` method

```python
df_india_json = pd.read_json('work_directory/pandas/output/survey_results_public_india_list-like.json', orient='records', lines=True)
```

```python
df_india_json.head()
```
