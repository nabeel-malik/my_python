```python
import pandas as pd
```

```python
df_v2 = pd.read_excel("work_directory/summary_table_testing/summary.xlsx", 'v2')
df_v3 = pd.read_excel("work_directory/summary_table_testing/summary.xlsx", 'v3')
```

```python
df_v2.head()
```

```python
fix_filter_v2 = df_v2['Check-type'] == 'Fix'
fix_filter_v3 = df_v3['Check-type'] == 'Fix'
```

```python
df_v2_fixes = df_v2[fix_filter_v2]
df_v3_fixes = df_v3[fix_filter_v3]
```

```python
pd.set_option('display.max_rows', None)
errors_v2 = df_v2_fixes['Finding (observation or error)'].value_counts()
errors_v3 = df_v3_fixes['Finding (observation or error)'].value_counts()
```

```python
df_fixes = pd.concat([errors_v2, errors_v3], axis=1)
```

```python
df_fixes.columns = ['v2', 'v3']
df_fixes
```
