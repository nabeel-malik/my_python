## Creating the first time series plot: `plt.plot_date([dates_list], y, linestyle=)`

Time series plots are simply plots of data against time (usually dates).

`plt.plot_date()` API: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot_date.html#matplotlib.pyplot.plot_date

Let's just plot a variable, `y`, against some dates:

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates


# Next 2 lines added to register matplotlib converters so that when pandas is imported, it overwrites matplotlib's built-in 
# datetime plotting with pandas datetime plotting.
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
 

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()
```

## Date formatting for time series plots

### Auto-formatting the ALIGNMENT of the date tick-labels on the current figure: `plt.gcf().autofmt_xdate()`

`plt.gcf().autofmt_xdate()` will rotate the tick-labels for the dates, and change their alignment, so that they fit a bit nicer.

The `.autofmt_xdate()` method is called on the current figure (`plt.gcf()`) and NOT directly on the `pyplot` object (`plt`).

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

# To rotate/align dates better we apply the following method on the current figure, instead of the pyplot object
plt.gcf().autofmt_xdate()              # '.gcf()' => get current figure

plt.tight_layout()
```

### FORMATTING THE DATES of the date tick-labels to, for instance 'May, 25 2019' like format: `plt.gca().xaxis.set_major_formatter(date_format)` where `date_format = mpl_dates.DateFormatter('%b, %d %Y')`

So, let's say we want to format the dates to a 'May, 25 2019' like format.

To do this we will use date/time formatting. 

To do this, we have already added `from matplotlib import dates as mpl_dates` at the top. From that imported module we are going to use the `DateFormatter()` class, and would pass a format code (string) to it. These are the same format codes that are used in the `.strftime()` method from the `datetime` class (from the `datetime` module). See the link below for these format codes.

`strftime()` and `strptime()` Format Codes: https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

Add link to \*.py for the `datetime` module explanation here. 

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

# To rotate/align dates better we apply the following method on the current figure, instead of the pyplot object
plt.gcf().autofmt_xdate()                                 # '.gcf()' => get current figure


# Formatting dates of the date tick-labels on the x-axis
date_format = mpl_dates.DateFormatter('%b, %d %Y')        # DateFormatter() class from the dates module in matplotlib library
plt.gca().xaxis.set_major_formatter(date_format)          # '.gca()' => get current axis

plt.tight_layout()
```

## Let's plot some real world data for daily Bitcoin prices

```python
plt.style.use('seaborn')

data = pd.read_csv('work_directory/matplotlib/code_snippets/08-TimeSeries/data.csv', parse_dates=['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

# To rotate/align dates better we apply the following method on the current figure, instead of the pyplot object
plt.gcf().autofmt_xdate()                                 # '.gcf()' => get current figure

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
```
