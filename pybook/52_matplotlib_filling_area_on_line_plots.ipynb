## Initial line plot (without fills) for median developer salaries versus age: `plt.plot(ages, dev_salaries)` & `plt.plot(ages, py_salaries)`

Let's just make initial line plots for all developer salaries (`dev_salaries`) and python developer salaries (`py_salaries`) versus age (`ages`):

```python
import pandas as pd
from matplotlib import pyplot as plt

# read data file containing median salaries versus age, for all_devs, python, and javascript
data = pd.read_csv('work_directory/matplotlib/code_snippets/05-Fill_Betweens/data.csv')

# defining series for plotting
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

## Adding fill, with transparency, between `py_salaries` and y = 0: `plt.fill_between(ages, py_salaries, alpha=0.25)`

```python
# read data file containing median salaries versus age, for all_devs, python, and javascript
data = pd.read_csv('work_directory/matplotlib/code_snippets/05-Fill_Betweens/data.csv')

# defining series for plotting
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

# adding fill between py_salaries and y=0
plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

## Modifying fill to appear between `py_salaries` and `overall_median`: `plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)`

```python
# read data file containing median salaries versus age, for all_devs, python, and javascript
data = pd.read_csv('work_directory/matplotlib/code_snippets/05-Fill_Betweens/data.csv')

# defining series for plotting
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median is the median salary for all developers who answered the survey (regardless of age)
overall_median = 57287

# adding fill between py_salaries and overall_median
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

## Using conditional fills with additional arguments: `plt.fill_between(where=, interpolate=, color=, label=)`

So let's create a:
- <font color=red>red</font> fill when `py_salaries` are <font color=red>below</font>  `overall_median`
- <font color=green>green</font> fill when `py_salaries` are <font color=green>above</font> `overall_median`

```python
# read data file containing median salaries versus age, for all_devs, python, and javascript
data = pd.read_csv('work_directory/matplotlib/code_snippets/05-Fill_Betweens/data.csv')

# defining series for plotting
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median is the median salary for all developers who answered the survey (regardless of age)
overall_median = 57287

# adding conditional fills between py_salaries and overall_median
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), interpolate=True, color='green', label='Above overall median',
                 alpha=0.25)
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), interpolate=True, color='red', label='Below overall median',
                 alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```
