## Creating the first pie chart: `plt.pie()`

`plt.pie()` API: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

```python
from matplotlib import pyplot as plt
```

```python
plt.ion()               # turn the interactive mode ON, so we do not have to run plt.show()
```

```python
plt.style.use('fivethirtyeight')

labels = ['Apples', 'Bananas', 'Oranges', 'Blueberries', 'Dates']
slices = [60, 80, 45, 120, 20]
colors = ['red', 'yellow', 'orange', 'blue', 'black']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title("Fruit Inventory")
plt.tight_layout()
```

Let's use **HEX color codes** instead of the built-in colors:

```python
plt.style.use('fivethirtyeight')

labels = ['Apples', 'Bananas', 'Oranges', 'Blueberries', 'Dates']
slices = [60, 80, 45, 120, 20]

'''
color HEX values:
blue: #008fd5
red: #fc4f30
yellow: #fceb24
orange: #f5923a
black: #000000
'''

colors = ['#fc4f30', '#fceb24', '#f5923a', '#008fd5', '#000000']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title("Fruit Inventory")
plt.tight_layout()
```

## Exploring more arguments: `plt.pie(<slices>, labels=, explode=, shadow=, startangle=, autopct=, wedgeprops=)`

```python
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]          # explode 'Python' slice out by 10% of the radius to place emphasis on it


plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

plt.title("Language Popularity")
plt.tight_layout()
plt.show()
```
